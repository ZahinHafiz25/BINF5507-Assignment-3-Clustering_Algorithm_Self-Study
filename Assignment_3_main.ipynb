```python
# Loading the datasets, algorithms, and functions:
from sklearn.datasets import make_moons, make_circles, make_blobs        # For loading the synthetic datasets from sklearn
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering      # For loading the clustering algorithms from sklearn 
from sklearn.preprocessing import StandardScaler                         # For standardizing the synthetic datasets
import matplotlib.pyplot as plt                                          # For plotting the synthetic datasets
```

```python
# Defining parameters for the synthetic datasets: 
X0, y0 = make_moons(n_samples = 300, shuffle = True, noise = 0.05, random_state = 150)                        # Dataset: 1 (make_moons)
X1, y1 = make_blobs(n_samples = 300, centers = 3, cluster_std = [0.5, 1.0, 2.5], random_state = 150)          # Dataset: 2 (make_blobs)
X2, y2 = make_circles(n_samples = 300, shuffle = None, noise = 0.05, random_state = 150, factor = 0.5)        # Dataset: 3 (make_circles)
```

```python
# For loop to iterate through the datasets and apply the clustering algorithms (DBSCAN, KMeans, Agglomerative Clustering) on each dataset:

for i, (X, y) in enumerate([(X0, y0), (X1, y1), (X2, y2)]):     # [(3 synthetic datasets x 3 clustering algorithms) = 9 graphs in total]
    X = StandardScaler().fit_transform(X)                       # Standardization of the synthetic datasets to make sure that the clustering algorithms will work effectively.  
    DBSCAN_algorithm = DBSCAN(eps = 0.3, min_samples = 3)              # This DBSCAN algorithm will be applied to all the 3 synthetic datasets. Recommended parameter: min_samples = 3 (GeeksforGeeks, 2025)
    KMeans_algorithm = KMeans(n_clusters = 3, random_state = 150)      # This KMeans algorithm will be applied to all the 3 synthetic datasets.
    Agglomerative_algorithm = AgglomerativeClustering(n_clusters = 3)  # This Agglomerative Clustering (Heirarchical Clustering) algorithm will be applied to all the 3 synthetic datasets.
    DBSCAN_fit = DBSCAN_algorithm.fit(X)                               # To fit the DBSCAN algorithm to all the standardized synthetic datasets.
    KMeans_fit = KMeans_algorithm.fit(X)                               # To fit the KMeans algorithm to all the standardized synthetic datasets.
    Agglomerative_fit = Agglomerative_algorithm.fit(X)                 # To fit the Agglomerative Clustering algorithm to all the standardized synthetic datasets.
    plt.figure(figsize = (24, 16))                                           # Using Matplotlib to plot the datasets. 
    plt.subplot(3, 1, 1)                                                     # To plot the first subplot of the figure.
    plt.scatter(X[:, 0], X[:, 1], c = DBSCAN_fit.labels_, cmap = 'viridis')  # To plot the first subplot of the figure with the DBSCAN algorithm using Matplotlib.
    plt.title("DBSCAN Clustering on Dataset " + str(i + 1))                  # To add a title and axes labels to the first subplot. 
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend(["DBSCAN"], loc = "upper right")                                       # To add a legend to the first subplot. 
    plt.subplot(3, 1, 2)                                                              # To plot the second subplot of the figure.
    plt.scatter(X[:, 0], X[:, 1], c = KMeans_fit.labels_, cmap = 'viridis')           # To plot the second subplot of the figure with the KMeans algorithm using Matplotlib.
    plt.title("KMeans Clustering on Dataset " + str(i + 1))                           # To add a title and axes labels to the second subplot.
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend(["KMeans"], loc = "upper right")                                       # To add a legend to the second subplot.
    plt.subplot(3, 1, 3)                                                              # To plot the third subplot of the figure.
    plt.scatter(X[:, 0], X[:, 1], c = Agglomerative_fit.labels_, cmap = 'viridis')    # To plot the third subplot of the figure with the Agglomerative Clustering algorithm using Matplotlib.
    plt.title("Agglomerative Clustering on Dataset " + str(i + 1))                    # To add a title and axes labels to the third subplot.
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend(["Agglomerative Clustering"], loc = "upper right")                       # To add a legend to the third subplot.
    plt.tight_layout()                                                                # To adjust the layout of the figure.
    plt.show()                                                                        # To show the figure with all the subplots. 
```
